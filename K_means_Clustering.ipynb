#***Introduction-***
K-means clustering is an Unsupervised machine learning algorithm. Here similar rows are clustered into groups.
Steps followed in K-means algorithm are as follows-
WhatsApp Image 2022-07-03 at 3.51.56 PM.jpeg

We can see the clusters formed by 2D data on our own, but we cant visualize the data in higher dimensions. This is what makes it powerful.

K-mean is a very powerful algorithm becaue, only calculation we do, is that we calculate Euclidean distances between all the points and compare to get closest centroid and asign that point to that centroid making it part of that cluster.

### Steps followed :
1. Initialize random centroids.
2. Find Euclidean distances of every point from each cluster.
3. Depending on least diatcance of the points from a centroid, assign a centroid for each point.
4. Now that we have clusters, we find centroids for these clusters.( mean of X-cooerdinates of all the points , mean of y-coordinates for each point.)
5. Now we repeat from step 2.
6. Updation of centroid values will stop the position of centriods do not change in conecutive steps.

###How will we know, how many clusters are to be formed?
This number can be found through 'Elbow Method'.
In thi metho a graph is plotted. On x-axi, we have number of clusters. And, on y-axis we have WCSS (within cluter sum of squares.)
